# Cluster Formation (Segmentation and Extraction)

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,10):
    kmeans = KMeans(i)
    kmeans.fit(df)
    wcss_iter = kmeans.inertia_
    wcss.append(wcss_iter)

number_clusters = range(1,10)
plt.plot(number_clusters,wcss)
plt.title('The Elbow title')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
```

```python
EVModel = KMeans(n_clusters=5)
EVModel.fit(df)
```

```python
identified_clusters = EVModel.fit_predict(df)
identified_clusters
```

```python
data_with_clusters = df_prev.copy()
data_with_clusters['Clusters'] = identified_clusters
```

```python
data_with_clusters.head(20)
```

```python
data_with_clusters['Clusters'].value_counts()
```

```python
Segment_wise_mean = data_with_clusters[['AccelSec','TopSpeed_KmH','Range_Km','Efficiency_WhKm','FastCharge_KmH','PriceEuro','Clusters','Seats']].groupby('Clusters').mean()
```
